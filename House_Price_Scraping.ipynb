```python
import time
import csv
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas
import re

def house_scrap(house, writer):
    title = house.find("div",{"class":"info-panel"}).a.string
    place = house.find("div",{"class":"where"}).findAll("span")
    neighbourhood = place[0].string
    layout = place[1].string.strip()
    square = place[2].string.strip()

    zones = house.find("div",{"class":"con"}).findAll("a")
    district = zones[0].string
    if len(zones) > 1:
        area = zones[1].string
    else:
        area = ""

    details = house.find("div",{"class":"con"}).findAll("span")

    level = ""
    facing = ""
    built = ""
    for detail in details:
        item = detail.next_sibling.string.strip()
        if '层' in item:
            level = item
        elif '朝' in item:
            facing = item
        elif '建' in item:
            built = item

    price = house.find("div",{"class":"price"}).span.string
    price_per_sqm = house.find("div",{"class":"price-pre"}).string

    metro = house.find("div",{"class":"view-label left"}).findAll("span", {"class":"fang-subway-ex"})
    if len(metro) == 1:
        metro = metro[0].string.strip()
    else:
        metro = ""
        
    # data cleaning
    
    room = re.sub("室", "", re.search("[0-9]室" , layout).group(0))
    lv_room = re.sub("厅", "", re.search("[0-9]厅" , layout).group(0))
    square = pandas.to_numeric(re.sub("平", "", square))
    price_per_sqm = pandas.to_numeric([re.sub("元/平", "", price_per_sqm)])
    
    if re.search("^.层/" , level) is not None:
        type = re.sub("层/", "", re.search("^.层/" , level).group(0))
    else:
        type = ""
    floor = re.sub("层", "", re.search("[0-9]+层" , level).group(0))
    
    if not built != built:
        built = pandas.to_numeric([re.sub("年建", "", built)])

    if metro == "":
        subline = ""
        subdist = ""
    else:
        subline = re.sub("号线", "", re.search("[0-9]+号线" , metro).group(0))
        subdist = re.sub("米", "", re.search("[0-9]+米" , metro).group(0))
    
    square_10 = round(square, -1)
    
    info = [neighbourhood, layout, square, district, area, level, facing, built, metro, price, price_per_sqm, room, lv_room, type, floor, subline, subdist, square_10]
    # print(info)
    writer.writerow(info)

def page_scrap(url,writer):
    html = urlopen(url)
    bsObj = BeautifulSoup(html)
    infos = bsObj.find("div", {"class":"list-wrap"})
    for house in infos.findAll("li"):
        house_scrap(house, writer)
        
file = "shlj.csv"
url = "http://sh.lianjia.com/ershoufang/o1d"
n_pages = 2100

start = time.time()

with open(file, 'w', newline='', encoding='utf-8') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(["neighbourhood", "layout", "square", "district", "area", 
                     "level", "facing", "built", "metro", "price", "price_per_sqm",
                     "room", "lv_room", "type", "floor", "subline", "subdist", "square_10"])
#     for i in range(2100, 2001):
    for i in range(n_pages):
        print("Page", i+1, "| Record", 20*i+1, "~", 20*(i+1))
        page_scrap(url+str(i+1),writer)

end = time.time()
print(end - start)
```

```python
data['room'] = [re.sub("室", "", re.search("[0-9]室" , text).group(0)) for text in data['layout']]
data['lv_room'] = [re.sub("厅", "", re.search("[0-9]厅" , text).group(0)) for text in data['layout']]
data['square'] = [re.sub("平", "", text) for text in data['square']]
data['price_per_sqm'] = [re.sub("元/平", "", text) for text in data['price_per_sqm']]
data['type'] = [re.sub("层/", "", re.search("^.层/" , text).group(0)) if re.search("^.层/" , text) is not None else "" for text in data['level']]
data['floor'] = [re.sub("层", "", re.search("[0-9]+层" , text).group(0)) for text in data['level']]
data['built'] = [re.sub("年建", "", text) if not text != text else "" for text in data['built']]
data['subline'] = [re.sub("号线", "", re.search("[0-9]+号线" , text).group(0)) if not text != text else "" for text in data['metro']]
data['subdist'] = [re.sub("米", "", re.search("[0-9]+米" , text).group(0)) if not text != text else "" for text in data['metro']]
```
